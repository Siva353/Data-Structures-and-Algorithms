```python
# 1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=6.
## Method definition
## Time complexity : O(3^n)
## Space complexity : O(n)
def T(n):
    ## Base case
    if n == 0:
        return 6
    ## recursive call
    else:
        return 3 * T(n-1) + 12 * n
    
print(T(2))

# T(n) = 3T(n-1) + 12n -> eq1
# let n = n-1 in eq1
# T(n-1) = 3T(n-2) + 12(n-1) -> eq2
# let n = n-2 in eq1
# T(n-2) = 3T(n-3) + 12(n-2) -> eq3
# now eq2 in eq1
# T(n) = 3[3T(n-2) + 12(n-1)] + 12n
# T(n) = 3^2T(n-2) + 3*12(n-1) + 12n -> eq4
# now eq3 in eq4 
# T(n) = 3^2[3T(n-3) + 12(n-2)] + 3*12(n-1) + 12n
# T(n) = 3^3T(n-3) + 3^2*12(n-2) + 3*12(n-1) + 12n ->eq5
# for repeated in K times
# T(n) = 3^kT(n-k) + 3^(k-1)*12(n-(k-1)) + 3^(k-2)*12(n-(k-2)) + ....... + 12n
# For base case n-k = 0 => n = k
# T(n) = 3^nT(n-n) + 3^2*12(n-n+1) + 3*12(n-n+2) + ...... + 12n
#      = 3^nT(0) + 3^2*12(1) + 3*12(2) + ...... + 12n
#      = 3^n series { It is in GP series}
#      = Time complexity is O(3^n)
```
